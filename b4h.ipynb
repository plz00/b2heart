```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("heart_diseasesss.csv")
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),cmap='YlGnBu',annot=True)
```

```python
sns.lineplot(data=df,x=df.age,y=df.cp,hue='target')
```

```python
sns.histplot(data=df,x=df.target,hue=df.sex)
```

```python
df.columns
```

```python
temp_df=df[['age','cp','trestbps','fbs','oldpeak','target']]
plt.figure(figsize=(15,10))
sns.pairplot(temp_df,hue='target')
plt.title('looking for Insites in Data')
plt.legend('heartdisease')
plt.tight_layout()
plt.plot()
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(temp_df.columns,1):
    plt.title(f"distrubtion{col}")
    sns.histplot(df[col],kde=True)
    plt.tight_layout()
    plt.plot()
```
