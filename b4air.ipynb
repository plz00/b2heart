```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AirQualityUCI.csv")
```

```python
df
```

```python
df=df.rename (columns={'T': 'Temperature'})
df=df.rename(columns={'RH': 'Relative Humidity'})
df=df.rename(columns={'AH': 'Absolute Humidity'})
df
```

```python

df['CO(GT)'] = pd.to_numeric(df['CO(GT)'].astype(str).str.replace(',', '.'), errors='coerce')
df['C6H6(GT)'] = pd.to_numeric(df['C6H6(GT)'].astype(str).str.replace(',', '.'), errors='coerce')
df['Temperature'] = pd.to_numeric(df['Temperature'].astype(str).str.replace(',', '.'), errors='coerce')
df['Relative Humidity'] = pd.to_numeric(df['Relative Humidity'].astype(str).str.replace(',', '.'), errors='coerce')
df['Absolute Humidity'] = pd.to_numeric(df['Absolute Humidity'].astype(str).str.replace(',', '.'), errors='coerce')
df
```

```python
df.dropna().sum()
```

```python
df['Absolute Humidity']=df['Absolute Humidity'].multiply(100)
df
```

```python
plt.figure(figsize=(10, 6))
numerical_df = df.select_dtypes(include=np.number)  
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
